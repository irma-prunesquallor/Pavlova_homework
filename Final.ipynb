## 1.2 Экономика игры

1. Match-3 позволяет получать XP, а также ресурсы: еду, железо. Кроме того, можно получить рекрутов, материалы и эмблемы. Требует затрат энергии (трех видов: мир / рейд / титан). Подряд можно пройти лишь несколько уровней, после чего энергия заканчивается.
2. Строительство зданий требует ресурсов: железа, еды. Значительные затраты времени
3. В свою очередь, здания позволяют фармить ресурсы: еду, железо
4. Тренировка героев: использует рекрутов, еду, материалы. Дает героев
5. Тренировка отрядов: требует затрат еды
6. Оптимизация команды героев: требует затрат еды и использования дубликатов героев
7. Призыв героев: требует затрат гемов, дает новых героев (возможны дубликаты) и отряды.
8. Ожидание: дает энергию, еду и железо.

Согласно сайту Deconstructor of Fun (Tuerlings, 2019), один из основных мотиваторов первичной конверсии игроков в разряд платящих - строительство зданий, поскольку оно требует значительных затрат времени. Это создает потребность в том, чтобы тратить внутриигровую валюту для ускорения и покупать VIP-pass, чтобы иметь возможность строить второе здание.

## 1.3 Гипотезы

Просмотр видеозаписей с игры позволяет сформулировать следующие гипотезы.

1. Если игрокам с большей длительностью загрузки игры чаще предоставлять скидки, то это может увеличить конверсию неплатящих игроков в платящие. Сейчас на видео видно, что длительность загрузки игры у неплатящего игрока выше, чем у платящего, что может свидетельствовать о более старой модели смартфона, а это, в свою очередь, о более низкой готовности платить.
2. Неплатящему игроку довольно рано приходят сообщения о поражении в PvP. Если дать неплатящему игроку больше времени не нести потери от вражеских рейдов, это может увеличить retention за счет снижения раздражения у игрока, у которого и так не много ресурсов.

Несколько гипотез на основе материалов в открытом доступе.

1. Если в качестве вознаграждения за просмотры рекламных роликов в игре использовать дополнительную возможность вызова героя, то это повысит число просмотренный роликов, поскольку игроки будут более заинтересованы в вознаграждении, которое можно сразу использовать, а не накапливать, а также игрокам захочется воспользоваться дополнительным шансом получить редкого героя (Tuerlings, 2019)
2. Если добавить цели на уровни или блокеры, то увеличится retention среди игроков, поскольку механика match-3 со временем надоедает игрокам, а в Empires & Puzzles она повторяется довольно много (добавление вариаций такого вида к основной механике - главные предложения по совершенствованию игры по данным GameRefinery, согласно (Tuerlings, 2019)).
3. Если увеличить доступный запас энергии, увеличив и время ее восполнения до 7-8 часов, возрастет retention, поскольку сейчас время восполнения энергии (3-4 часа) и время, на которое ее хватает, не совпадает со стандартными циклами в жизни обычного игрока - 7-8 часов сон, 8 часов рабочий день (Мудрый Гик, 2017)

#### Средний баланс гемов

Рассчитаем __средний баланс гемов__ по дням для неплатящего и платящего игрока.

```python
# построим частотную диаграмму
avg_gem['Non-Paying'].plot(kind='hist', 
                     title='Средний баланс гемов в день',
                     alpha = 0.5,
                      bins=10,
                     density = True)


avg_gem['Whale'].plot(kind='hist', 
                     title='Средний баланс гемов в день',
                     alpha = 0.5,
                      bins=10,
                     density = True)
```

## 2.2-2.4. Анализ метрик, визуализация, проверка отличий на статистическую значимость

### 1) Длина сессии

Мы знаем, что общее время, проведенное в игре, у платящего игрока и неплатящего игрока примерно одинаковое, однако неплатящий игрок за 14 дней сыграл больше сессиий, чем платящий. Таким образом, неплатящий игрок заходит в игру чаще, но на более короткие промежутки времени - иными словами, средняя длина одной сессии неплатящего игрока должна быть меньше.
Чтобы проверить это, формируем таблицы для платящего и неплатящего игрока, где просуммировано время каждой сессии.

```python
#у неплатящего игрока
time_sess_np = df_non_paying_short.groupby(['Session']).agg(['sum'])['Time']
time_sess_np.rename(columns={'sum': 'Non-paying'}, inplace=True)
#переведем время в секунды
time_sess_np['Non-paying_sec'] = time_sess_np['Non-paying'] / np.timedelta64(1, 's')
time_sess_np.head()
```

```python
#у платящего игрока
time_sess_p = df_whales_short.groupby(['Session']).agg(['sum'])['Time']
time_sess_p.rename(columns={'sum': 'Whale'}, inplace=True)
#переведем время в секунды
time_sess_p['Whale_sec'] = time_sess_p['Whale'] / np.timedelta64(1, 's')
time_sess_p.head()
```

```python
# посмотрим, сколько в среднем длится сессия у каждого из игроков
print(time_sess_np['Non-paying_sec'].mean(), time_sess_p['Whale_sec'].mean())
```

```python
# построим частотную диаграмму
time_sess_np['Non-paying_sec'].plot(kind='hist', 
                     title='Длина сессии',
                     alpha = 0.5,
                      bins=16,
                     density = True)


time_sess_p['Whale_sec'].plot(kind='hist', 
                     title='Длина сессии',
                     alpha = 0.5,
                      bins=16,
                     density = True)
```

На графике видно, что распределение у неплатящего игрока смещено влево, что соответствует нашим первоначальным представлениям.

По крайней мере у неплатящего игрока распределение не похоже на нормальное, в связи с чем t-тест навряд ли применим. Тем не менее, попробуем использовать его для сравнения средней длины сессии у неплатящего и платящего игроков.

Сделаем оговорку, что в случае наличия большего объема данных было бы правильно собрать поюзерную статистику, но в данном случае у нас есть сведения только о двух игроках. По смыслу мы анализируем не среднего игрока, а среднюю сессию двух конкретных игроков, что не совсем соответствует поставленной задаче (выявить отличия платящих и неплатящих), но будем использовать те данные, которые у нас есть.

```python
import scipy
from scipy import stats
scipy.stats.ttest_ind(time_sess_np['Non-paying_sec'], time_sess_p['Whale_sec'])
```

На уровне значимости 10% мы можем утверждать, что гипотеза о равенстве средней длины сессии между неплатящими и платящими игроками отвергается. Иными словами, значимые (на уровне значимости 10%) различия между длиной сессии у платящего и неплатящего игрока все же есть. Однако данный результат ненадежен, поскольку условия применимости t-теста по всей видимости не выполнены.

Попробуем применить ранговый тест Манна-Уитни, который не требует нормального распределения величин.

```python
scipy.stats.mannwhitneyu(time_sess_np['Non-paying_sec'], time_sess_p['Whale_sec'])
```

По итогам данного теста, даже на 5%-м уровне значимости мы можем утверждать, что длина сессии и платящего и неплатящего игрока отличается.

Мы использовали двусторонний тест и подтвердили, что величины значимо отличаются друг от друга, но не уточнили, в какую сторону. Исходя из нашей логики, длина сессии у неплатящего игрока должна быть меньше, чем у платящего. Для того чтобы проверить это, используем односторонний тест, где альтернативная гипотеза - что длина сессии у неплатящего игрока стохастически меньше, чем у платящего.

```python
scipy.stats.mannwhitneyu(time_sess_np['Non-paying_sec'], time_sess_p['Whale_sec'], alternative='less')
```

На 5%-м уровне значимости мы можем утверждать, что длина сессии у неплатящего игрока меньше, чем у платящего.

## 2.6 Гипотезы

С учетом изложенных выше ограничений, попробуем все же сформулировать первую гипотезу на основе проведенного анализа, а вторую гипотезу сформулируем на основе визуального анализа данных. 

Допустим, что, несмотря на ограничения проведенного анализа, мы все же опираемся на полученный результат относительно активности в битвах с титанами в составе альянса: в среднем в день платящий и неплатящий игрок значимо не отличаются по склонности участвовать в таких битвах. Можно предположить, что в таком случае можно сделать из этого действия платную фичу: например, установить небольшой ценник за саму возможность вступать в альянсы (и, соответственно, биться с титанами в их составе, а также совершать другие действия). Мы знаем, что неплатящие игроки ценят эту функцию в игре, как и платящие, поэтому можем попробовать сделать ее платной. Предположительно, для уже платящих игроков это не станет проблемой и не приведет к их отказу от игры, а неплатящих может сподвигнуть совершить покупку.

Такое предложение также согласуется с результатами исследований о том, за какие возможности в игре игроки готовы платить. Присоединяясь к альянсу, помимо возможности объединения с другими игроками, что тоже ценно, игрок приобретает дополнительную возможность и соревноваться с другими игроками - соревноваться против других альянсов. Возможность соревноваться с другими - второй по популярности ответ "китов" на вопрос о том, какая возможность в игре заставляет из платить, а для платящих не слишком большие суммы данный ответ четвертый по популярности согласно исследованию Newzoo (2014).

Результат можно усилить с помощью эндаумент-эффекта, если давать игрокам сначала попробовать присоединение к альянсу, а по истечении некоторого срока устанавливать плату за эти возможности. Мы видели, что первый "пик" количества битв в день у платящего игрока приходится на второй день, а у неплатящего игрока - на 7-й день, то есть ему требуетя больше времени на то, чтобы оценить эту возможность. Исходя из этого, срок бесплатного доступа к альянсам можно установить на уровне 7 дней.

__Гипотеза 1__: _Если для всех новых игроков взимать плату за возможность вступления в альянс, начиная с 8 дня игры, то увеличится конверсия неплатящих игроков в платящих (нет данных для того, чтобы предположить, на сколько), поскольку в настоящий момент без такой платы и платящие, и неплатящие игроки активно участвуют в альянсах и битвах, и именно к 8 дню данную возможность успевают по достоинству оценить неплатящие игроки._

Проанализируем поведение неплатящего игрока, когда он заходит в магазин (Tag = "Shop", Action = "Start"). В столбце Info во время восьмой сессии отмечено, что в определенный момент игрок испытывает нехватку ресурса "еда": "еды мало, но не хватает премиума, чтобы купить её в магазине". Во время десятой сессии также есть отметка "думала купить еды, но оказалось дорого". Если мы считаем, что поведение данного игрока репрезентативно относительно поведения всех остальных неплатящих игроков, можем предположить, что в аналогичной ситуации другим неплатящим игрокам тоже начинает не хватать еды. Тот факт, что на первых порах в игре действительно часто не хватает ресурсов (еды, железа) подтверждается и в обзорах игры (например, на сайте "Мудрый Гик": https://wisegeek.ru/aac/empires-puzzles#10). Вероятно, в эти моменты в игре (например, на 8-й или 10-й сессии) можно было бы попробовать давать специальное предложение в магазине: скидку (допустим, на 20%) на базовый набор из 400 гемов. Предполагаем, что это может подтолкнуть неплатящего игрока перешагнуть порог первой покупки. Выбираем 10-ю сессию в предположении, что к этому моменту игрок осознал, что быстро набрать достаточно гемов для решения проблемы без оплаты у него не получается.

Соответственно, гипотеза 2 будет звучать следующим образом.

__Гипотеза 2__: _Если для неплатящих игроков во время 10-й сессии предложить набор из 400 гемов со скидкой 20%, то конверсия неплатящих пользователей в платящих повысится (нет данных, чтобы предположить, на сколько), поскольку на данном этапе неплатящие игроки начинают испытывать нехватку ресурсов - достаточную для того, чтобы проверить предложения в магазине, но недостаточную для того, чтобы совершить покупку при текущих ценах._

# Заключение

С учетом описанных ограничений ясно, что сформулированные гипотезы по улучшению геймдизайна и повышению вероятности конверсии неплатящих игроков в платящих не жестко привязаны к полученным выводам из количественного анализа. Хотя при этом частично гипотезы подтверждаются внешней аналитикой, для уточнения гипотез и выработки новых потребуются данные в  большем объеме и по выборкам платящих и неплатящих игроков.
