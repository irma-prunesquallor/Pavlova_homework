# Задание 2
Добавьте в класс Rate параметр diff (со значениями True или False), который в случае значения True в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением. Считайте, self.diff будет принимать значение True только при возврате значения курса. При отображении всей информации о валюте он не используется.

```python
class Rate:
    def __init__(self, format_='value', diff = False):
        self.format = format_
        self.diff = diff
            
    def exchange_rates(self):
        """
        Возвращает ответ сервиса с информацией о валютах в виде:
        
        {
            'AMD': {
                'CharCode': 'AMD',
                'ID': 'R01060',
                'Name': 'Армянских драмов',
                'Nominal': 100,
                'NumCode': '051',
                'Previous': 14.103,
                'Value': 14.0879
                },
            ...
        }
        """
        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
    
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate('full').make_format('EUR')
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        
        Rate('value').make_format('EUR')
        79.4966
        """
        self.response = self.exchange_rates()
        
        if currency in self.response:
            if self.format == 'full':
                return self.response[currency]
            
            if self.format == 'value':
                return self.response[currency]['Value']
        
        return 'Error'
    
    def eur(self, diff):
        """Возвращает курс евро на сегодня в формате self.format или в виде разницы между текущим и прошлым значением"""
        if diff == False:
            return self.make_format('EUR')
        else:
            return (self.response['EUR']['Value'] - self.response['EUR']['Previous'])
    
    def usd(self, diff):
        """Возвращает курс доллара на сегодня в формате self.format или в виде разницы между текущим и прошлым значением"""
        if diff == False:
            return self.make_format('USD')
        else:
            return (self.response['USD']['Value'] - self.response['USD']['Previous'])       
    
    def brl(self, diff):
        """Возвращает курс бразильского реала на сегодня в формате self.format или в виде разницы между текущим и прошлым значением"""
        if diff == False:
            return self.make_format('BRL')
        else:
            return (self.response['BRL']['Value'] - self.response['BRL']['Previous']) 
        
```

# Задание 3
Напишите класс Designer, который учитывает количество международных премий. Подсказки в коде занятия в разделе “Домашнее задание задача 3”.

Комментарий по классу Designer такой:
Напишите класс Designer, который учитывает количество международных премий для дизайнеров (из презентации: “Повышение на 1 грейд за каждые 7 баллов. Получение международной премии – это +2 балла”). Считайте, что при выходе на работу сотрудник уже имеет две премии и их количество не меняется со стажем (конечно если хотите это можно вручную менять).

Класс Designer пишется по аналогии с классом Developer из материалов занятия. Комментарий про его условия Вика написала выше: “Повышение на 1 грейд за каждые 7 баллов. Получение международной премии – это +2 балла”

```python
class Employee:
    def __init__(self, name, seniority):
        self.name = name
        self.seniority = seniority
        
        self.grade = 1
            
    def grade_up(self):
        """Повышает уровень сотрудника"""
        self.grade += 1
    
    def publish_grade(self):
        """Публикация результатов аккредитации сотрудников"""
        print(self.name, self.grade)
```

```python
class Designer(Employee):
    def __init__(self, name, seniority):
        super().__init__(name, seniority)
        # сколько баллов приносит каждая международная премия - предполагаем, что +2
        self.award_bonus = 2
        # сколько премий есть у дизайнера - предполагаем что на старте есть уже 2
        self.award_number = 2
    
    def check_if_it_is_time_for_upgrade(self):
        # для каждой аккредитации увеличиваем счетчик на 1
        # считаем, что все дизайнеры проходят аккредитацию
        self.seniority += 1
        
        # условие повышения сотрудника из презентации
        if ((self.award_bonus * self.award_number) + self.seniority) % 7 == 0:
            self.grade_up()
        
        # публикация результатов
        return self.publish_grade()
```

```python
alex = Designer('Александр', 0)
```

```python
for i in range(20):
    alex.check_if_it_is_time_for_upgrade()
```
