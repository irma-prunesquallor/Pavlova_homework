# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
# читаем таблицу с рейтингами, смотрим как выглядит
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# вводим функцию, присваивающую рейтинг фильму. Для упрощения высокий рейтинг - все, что выше 4
def rating_class(rating):
    if rating <= 2:
        return "низкий рейтинг"
    elif 2 < rating <= 4:
        return "средний рейтинг"
    else:
        return "высокий рейтинг"
```

```python
# создаем столбец class с присвоенной категорией рейтинга
ratings['class'] = ratings['rating'].apply(rating_class)
```

```python
# проверяем, как выглядит
ratings.head()
```
