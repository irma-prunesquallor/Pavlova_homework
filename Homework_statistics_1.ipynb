Найдем выбросы для температуры. Экспертно тут сложно что-то сказать, воспользуемся IQR.

```python
q1_rt = horse_data['rectal_temperature'].quantile(0.25)
q3_rt = horse_data['rectal_temperature'].quantile(0.75)
iqr_rt = q3_rt - q1_rt
lower_bound_rt = q1_rt - (1.5 * iqr_rt) 
upper_bound_rt = q3_rt + (1.5 * iqr_rt)
print('Нижнее нормальное значение:', lower_bound_rt)
print('Верхнее нормальное значение:', upper_bound_rt)
horse_data.loc[(horse_data['rectal_temperature'] < lower_bound_rt) | (horse_data['rectal_temperature'] > upper_bound_rt)]
```

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

Эта же строка - единственный случай отсутствия данных по outcome.
Если цель нашего исследования - понять, как операция влияет на результаты лечения или другой показатель, то можно было бы вообще удалить эту строку.
Пока специальной цели такого рода нет, можем заполнить отсутстующие данные по операции и по исходу модальными значениями.

```python
horse_data['surgery?'].fillna(horse_data['surgery?'].mode()[0], inplace=True)
horse_data['outcome'].fillna(horse_data['outcome'].mode()[0], inplace=True)
```

Посмотрим, что из себя представляет  отсутствие данных по температуре тела.

```python
horse_data.loc[horse_data['rectal_temperature'].isnull()]
```

Наблюдения, где нет сведений о температуре, часто характеризуются отсутствием и других сведений о состоянии лошади.
Возможно, это были какие-то срочные осмотры и операции; или проблема была очевидна; или данные не успели внести в базу.
Помним, что rectal_temperature имела довольно много выбросов, в связи с чем пропуски целесообразно заполнять медианным значением.
В описании данных указано, что температура тела и температура конечностей могут быть связаны. В связи с этим, возможно, что заполнять rectal_temperature стоит медианным значением, группируя по temperature_of_extremities.

```python
print(horse_data.groupby('temperature_of_extremities')['rectal_temperature'].median())
```

Однако разница в медианных значениях при такой группировке получилась незначительная. Поэтому можем заполнить просто медианой, без группировки.

```python
horse_data['rectal_temperature'].fillna(horse_data['rectal_temperature'].median(), inplace=True)
```

Температура конечностей также может отражать состояние шока, поэтому медианный пульс может отличаться в зависимости от температуры конечностей. Проверим.

```python
print(horse_data.groupby('temperature_of_extremities')['pulse'].median())
```

Медианный пульс действительно отличается довольно существенно, поэтому заполним столбец с пульсом медианой, группируя по температуре конечностей.

```python
horse_data['pulse'].fillna(horse_data.groupby(['temperature_of_extremities'])['pulse'].transform('median'), inplace=True)
```

Посмотрим, какие значения пульса остались незаполненными.

```python
horse_data.loc[horse_data['pulse'].isnull()]
```

Остались незаполненными значения, где нет информации по температуре конечностей. Тогда на следующем этапе заполним temperature_of_extremities.
Учитывая связь с пульсом, там, где пульс известен, мы можем точнее восстановить значение температуры конечностей, чем просто мода.
Согласно описанию данных, нормальный пульс - 30-40, допустимо 20-25. Учащенный пульс может соответствовать состоянию шока, а для этого состояния характерны значения temperature_of_extremities на уровне 3 или 4.
Для упрощения (тут нужна экспертная оценка, мы же интуитивно оттолкнемся от медиан: 54 и 56 для нормальной и повышенной температуры конечностей, 80 и 90 для пониженной) примем учащенным пульс выше 70, и соответствующую ему temperature_of_extremities 3, чтобы меньше повлиять на моду (мода равна 3), а пульсу ниже 70 пусть соответствует нормальная температура 1. (Это довольно произвольно.). Строки, где данных по пульсу нет, тоже заполним модой

```python
for pulse in horse_data['pulse']:
    if pulse <= 70:
        horse_data['temperature_of_extremities'].fillna('1', inplace=True)
    else: 
        horse_data['temperature_of_extremities'].fillna('3', inplace=True)
```

```python
horse_data['temperature_of_extremities'].mode()[0]
```

Изменилась модальное значение температуры конечностей. Возможно, мы исказили картину по выборке, выбрав неправильный способ заполнения пробелов в данных. С другой стороны, это может быть признаком того, что о нормальных значениях температуры конечностей данные в карту просто не заносились как неинформативные для диагностики.

Оставшиеся пустые значения пульса заполняем медианой.

```python
horse_data['pulse'].fillna(horse_data['pulse'].median(), inplace=True)
```

Заполним пропуски в уровне боли. Уровень боли тоже связан с пульсом. Но, если судить по медианам, пульс не монотонно связан с оценкой уровня боли.

```python
print(horse_data.groupby('pain')['pulse'].median())
```

Модальный уровень боли 3. Им и заполним пропущенные значения, но если пульс низкий (будем считать, что меньше 50), то оценим боль на 1, а если слишком высокий (будем считать, что выше 85), оценим на 5. 

```python
for pulse in horse_data['pulse']:
    if pulse <= 50:
        horse_data['pain'].fillna('1', inplace=True)
    elif pulse >= 85:
        horse_data['pain'].fillna('5', inplace=True)
    else: 
        horse_data['pain'].fillna('3', inplace=True)
```

```python
horse_data['pain'].mode()[0]
```

Модальное значение уровня боли не изменилось.

Осталось заполнить пропущенные значения respiratory_rate. Можно предположить, что частота дыхания связана с пульсом, температурой, уровнем боли. Проверим последнее.

```python
print(horse_data.groupby('pain')['respiratory_rate'].median())
```

Связь не совсем очевидна. Кроме того, в описании данных указано, что данный показатель слабо информативен ("usefulness is doubtful due to the great fluctuations"). Учитывая это, и учитывая наличие многочисленных выбросов, заполним пропуски медианным значением.

```python
horse_data['respiratory_rate'].fillna(horse_data['respiratory_rate'].median(), inplace=True)
```

Проверим, что нет пропусков.

```python
horse_data.describe(include='all')
```

По показателю count везде имеется 300 непустых значений, значит, пропуски заполнены.
